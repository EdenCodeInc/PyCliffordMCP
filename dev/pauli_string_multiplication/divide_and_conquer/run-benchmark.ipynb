# Divide-and-Conquer Pauli String Multiplication Benchmark

This notebook runs the simplified divide-and-conquer benchmark for Pauli string multiplication.


```python
# Setup and Import
import os
os.chdir('/Users/hwanda/Desktop/EdenCode/projects/PyCliffordMCP-main/dev/Pauli_string_dev/dev_test_pauli_string_divide_and_conquer')

import llm_divide_conquer_benchmark as benchmark
from itertools import product

# Show current model configuration
print(f"Current Model: {benchmark.MODEL_NAME}")
print(f"Backend: {benchmark.MODEL_BACKEND}")
print(f"Save LLM Responses: {benchmark.SAVE_LLM_RESPONSES}")
```

```python
# Model configuration
benchmark.MODEL_BACKEND = "gemini"
benchmark.MODEL_NAME = "gemini-2.5-pro-preview-06-05"

# Parameter ranges - Small test for debugging format
N_values = [150,200]              # Problem sizes (number of qubits)
chunk_size_values = [2,5,20]      # Chunk sizes for divide-and-conquer
batch_sizes = [10]           # Number of problems per batch (smaller for easier debugging)
iterations = [10]            # Number of iterations per configuration
PHASE_ACCUMULATION = [True] # True: Accumulate phases, False: Don't accumulate phases
LLM_RESPONSES = [True] # True: Save LLM responses, False: Don't save LLM responses

# Run all parameter combinations
all_results = []

for phase_accumulation, llm_responses in product(PHASE_ACCUMULATION, LLM_RESPONSES):
    benchmark.USE_PHASE_ACCUMULATION = phase_accumulation
    benchmark.SAVE_LLM_RESPONSES = llm_responses

    for N, chunk_size, batch_size, num_iterations in product(N_values, chunk_size_values, batch_sizes, iterations):
        if chunk_size >= N:
            print(f"Skipping N={N}, chunk_size={chunk_size} (chunk too large)")
            continue

        print(f"\n=== Running: N={N}, chunk_size={chunk_size}, batch_size={batch_size}, iterations={num_iterations} ===")

        # Run experiment for this configuration
        results = benchmark.run_experiment([N], [chunk_size], [batch_size], [num_iterations])
        all_results.extend(results)

        # Print summary for this configuration
        for result in results:
            print(f"  Success Rate: {result['success_rate']:.1%}")
            print(f"  Duration: {result['duration']:.1f}s")
            print(f"  LLM Calls: {result['llm_calls']}")
            print(f"  Record ID: {result['record_id']}")

# Final summary
print(f"\n=== COMPLETE: Ran {len(all_results)} experiments ===")
print(f"Records saved to: records/{benchmark.MODEL_NAME}/")
print("\nNew JSON format with problems_detailed for easy debugging!")
```
