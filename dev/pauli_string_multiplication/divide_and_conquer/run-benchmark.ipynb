# Divide-and-Conquer Pauli String Multiplication Benchmark

Simple notebook for running phase accumulation divide-and-conquer experiments.


```python
# Setup
import os
import sys
from itertools import product

# Navigate to benchmark directory
project_root = '/Users/hwanda/Desktop/EdenCode/projects/PyCliffordMCP'
benchmark_dir = os.path.join(project_root, 'dev/pauli_string_multiplication/divide_and_conquer')
os.chdir(benchmark_dir)
sys.path.insert(0, project_root)

import llm_divide_conquer_benchmark as benchmark
```

```python
# Set model configuration
benchmark.LLM_BACKEND = "gemini"  # or "openai" or "claude"

# Define parameter ranges
model_names = ["gemini-2.5-flash-preview-04-17-thinking"]
save_llm_responses = [True]
N_values = [7, 12]               # Problem sizes (number of qubits)
chunk_size_values = [2, 3]       # Chunk sizes for divide-and-conquer
batch_sizes = [3]               # Number of problems per batch
num_iterations_list = [3]       # Number of iterations per configuration
temperature_values = [0.0]      # LLM temperature values

# Run all combinations
for model_name, save_llm_response, N, chunk_size, batch_size, num_iterations, temperature in product(
    model_names, save_llm_responses, N_values, chunk_size_values, batch_sizes, num_iterations_list, temperature_values):
    
    if chunk_size >= N:
        print(f"Skipping N={N}, chunk_size={chunk_size} (chunk too large)")
        continue
        
    # Set global configuration
    benchmark.MODEL_NAME = model_name
    benchmark.SAVE_LLM_RESPONSES = save_llm_response
    benchmark.TEMPERATURE = temperature
    
    print(f"\nRunning: N={N}, chunk_size={chunk_size}, batch_size={batch_size}, iterations={num_iterations}, temp={temperature}")
    
    # Run experiment
    results = benchmark.run_experiment([N], [chunk_size], [batch_size], [num_iterations], [temperature])
    
    # Show results
    for result in results:
        print(f"  Accuracy: {result['accuracy']:.1%}")
        print(f"  Duration: {result['duration']:.1f}s")
        print(f"  LLM Calls: {result['llm_calls']}")
        print(f"  Record: {result['record_id']}")
    
    print("Finished.\n")
```
