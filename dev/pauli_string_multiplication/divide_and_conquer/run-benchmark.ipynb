# Divide-and-Conquer Pauli String Multiplication Benchmark

This notebook runs phase accumulation divide-and-conquer experiments using the same pattern as direct multiplication.


```python
# Setup
import os
import sys
from itertools import product

# Navigate to benchmark directory
project_root = '/Users/hwanda/Desktop/EdenCode/projects/PyCliffordMCP'
benchmark_dir = os.path.join(project_root, 'dev/pauli_string_multiplication/divide_and_conquer')
os.chdir(benchmark_dir)
sys.path.insert(0, project_root)

import llm_divide_conquer_benchmark as benchmark

print(f"Current directory: {os.getcwd()}")
print(f"Temperature: {benchmark.TEMPERATURE}")
```

```python
# Set model configuration
benchmark.LLM_BACKEND = "gemini"  # or "openai" or "claude"

# Define parameter ranges
model_names = ["gemini-2.5-flash"]
save_llm_responses = [False]
N_values = list(range(3,102,3))              # Problem sizes (number of qubits)
batch_sizes = [10]               # Number of problems per batch
num_iterations_list = [10]       # Number of iterations per configuration
temperature_values = [0.0]      # LLM temperature values

# Run all combinations
for model_name, save_llm_response, N, batch_size, num_iterations, temperature in product(
    model_names, save_llm_responses, N_values, batch_sizes, num_iterations_list, temperature_values):
    
    # Set global configuration
    benchmark.MODEL_NAME = model_name
    benchmark.SAVE_LLM_RESPONSES = save_llm_response
    benchmark.N = N
    benchmark.chunk_size = N//3
    benchmark.batch_size = batch_size
    benchmark.num_iterations = num_iterations
    benchmark.TEMPERATURE = temperature
    
    print(f"\nRunning: N={N}, chunk_size={N//3}, batch_size={batch_size}, iterations={num_iterations}, temp={temperature}")
    
    # Run experiment (just call main() like direct multiplication)
    benchmark.main()
    
    print("Finished.\n")
```
