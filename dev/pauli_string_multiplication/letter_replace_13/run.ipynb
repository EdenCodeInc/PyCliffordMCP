```python
# Run the letter-replace-13 benchmark using Anthropic Claude (most powerful)
from itertools import product
import letter_replace_benchmrk as bench

bench.LLM_BACKEND = "grok"

# Configure runs
model_names = [
    "grok-4-fast-reasoning"
]
save_llm_responses = [True]
L_values = [21,22,23,24,25,50,100,150,200,250,300]   # lengths to test
temperature_values = [0.0]      # Claude supports 0.0 (deterministic)
batch_sizes = [10]
num_iterations_list = [10]

for model_name, save_llm_response, L, temperature, batch_size, num_iterations in product(
    model_names, save_llm_responses, L_values, temperature_values, batch_sizes, num_iterations_list
):
    bench.MODEL_NAME = model_name
    bench.SAVE_LLM_RESPONSE = save_llm_response
    bench.L = L
    bench.TEMPERATURE = temperature
    bench.batch_size = batch_size
    bench.num_iterations = num_iterations

    print(f"\nRunning benchmark: model={model_name}, L={L}, temperature={temperature}, "
          f"batch_size={batch_size}, iterations={num_iterations}")
    bench.main()
    print("Finished.\n")
```
